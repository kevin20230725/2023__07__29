```python
#讀取csv檔，成為2維的list
import csv
with open("a2.csv",mode='r',encoding='utf-8',newline='') as file :
    csv_reader = csv.reader(file)
    next(csv_reader)        #跳一行
    #print(type(csv_reader))        此時type為csv.reader
    csv_list = list(csv_reader)
    #print(type(csv_list))
csv_list
```

```python
for student in csv_list:
    if student[0] == '李慈啟':
        print(student)
        break
```
