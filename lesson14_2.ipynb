```python
#matplotlib.pyplot
#參考網站:https://matplotlib.org/stable/api/pyplot_summary.html

import matplotlib.pyplot as plt


# 1.製作figure  (類似畫布)
fig = plt.figure()

# 2.在1的figure上製作一個axes
ax_1 = fig.add_subplot(1,1,1)

# 3.將圖的數據存在axes中
X = [0,1,2]
Y = [3,4,5]
ax_1.plot(X,Y)

# 4.顯示圖
plt.show()
```

```python
figure = plt.figure()
axes1 = figure.add_subplot(1,3,1)       # row : col = 1:3，然後1*3=3等分，所以最後位置總共可以填到3，最後的1代表第1個分割位置
axes1.set_title('plot 1')

axes1 = figure.add_subplot(1,3,2)
axes1.set_title('plot 2')

axes1 = figure.add_subplot(1,3,3)
axes1.set_title('plot 3')

plt.show()
```

```python
import matplotlib.pyplot as plt

data = [-1,-4.3,15,3,12]
#創建了一個 Figure 物件，指定了圖表的尺寸為 (3, 3)。Figure 物件是一個包含所有繪製內容的容器。
figure = plt.figure(figsize=(3,3))
#向 Figure 物件添加了一個子圖。在沒有指定參數的情況下，add_subplot() 預設會創建一個單一的子圖
axes = figure.add_subplot()
#使用 plot() 方法在子圖上繪製了數據。'r+' 參數表示線條的顏色為紅色，點的樣式為加號（+）。
#因為未指定 X 軸的數據，所以默認會將 data 列表的索引作為 X 軸的數據。
axes.plot(data,'r+')
```

```python
import matplotlib.pyplot as plt

student1 = {'chinese':74,'english':96,'math':100,'history':66,'discover':88}
student1.keys()
student1.values()

figure = plt.figure(figsize=(8,6),dpi=100,facecolor='#cccccc')     #dpi = 解析度(預設100)、facecolor = R(cc) G(cc) B(cc) 因為依樣所以底為灰色
axes1 = figure.add_subplot()
axes1.plot(student1.keys(),student1.values(),'r--+',label='stu1')       # r 是顏色(紅色)、--是線的表示方式、+是點的表示方式

student2 = {'chinese':88,'english':75,'math':66,'history':90,'discover':75}
axes1.plot(student2.values(),'b:D',label='stu2') 

axes1.legend()      #顯示label框框   
plt.show()
```
