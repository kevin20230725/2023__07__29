```python
# 目的:讓圖表下標轉為中文

#使用codespace 要安裝中文字型
#install  要加-y
#命令列:sudo apt-get update -y
#命令列:sudo apt-get install -y fonts-wqy-zenhei
#說明網址:https://askubuntu.com/questions/490829/how-can-i-install-chinese-fonts-on-kubuntu-14-04
#字型位置:/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
#mac  font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#mac  font1=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=12)
#window  font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)
font=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)

#先定義參數
student1 = {'國':80,'英':90,'數':95,'社':80,'自':85}
student2 = {'國':90,'英':90,'數':70,'社':85,'自':60}

student1_value_list = list(student1.values())
student2_value_list = list(student2.values())

#建立畫布
figure = plt.figure(figsize=[8,5],facecolor='#cccccc')
#建立圖表
axe = figure.add_subplot(1,1,1)
#作圖
axe.plot(student1.keys(),student1_value_list,'r--+',label='學生1')
axe.plot(student2.keys(),student2_value_list,label='學生2')
#將x軸設定為指定字體
axe.set_xticklabels(student1.keys(),fontproperties=font)
#只要是中文都要加fontproperties=font，否則無法顯示
#設定title
axe.set_title('學生的分數',fontproperties=font)
axe.set_xlabel('科目',fontproperties=font)
axe.set_ylabel('分數',fontproperties=font)
#加入格線
axe.grid(axis='y')
#加入標籤說明，prop=font 為顯示中文
axe.legend(prop=font)
#逐點顯示數值     axe.annotate()
#   axe.annotate(90,[0,90])
#step 1 先找x軸的標籤位置
xlocation = axe.get_xticks()
xtick_count = len(xlocation)
#step 2 利用迴圈逐點標記
for i in range(xtick_count):
    x = xlocation[i]
    y1 = student1_value_list[i]
    y2 = student2_value_list[i]
    axe.annotate(str(y1),[x,y1])
    axe.annotate(str(y2),[x,y2])
    #也可將以上寫成以下二行
    #axe.annotate(str(student1_value_list[i]),[i,student1_value_list[i]])
    #axe.annotate(str(student2_value_list[i]),[i,student2_value_list[i]])
    


plt.show()








```
