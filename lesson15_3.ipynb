```python
import numpy as np
import matplotlib.pyplot as plt
#建立一個等距數值
x_array = np.linspace(-3,3,20)
y1_array = x_array*2+1
y2_array = x_array**2

#建立畫布
figure = plt.figure(figsize=[8,5],facecolor='#cccccc')
#建立圖表
axes = figure.add_subplot(1,1,1)
#作圖ㄌ
axes.plot(x_array,y1_array)
axes.plot(x_array,y2_array)
plt.show
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
#建立等距序列
x = np.arange(start=0,stop=2.5,step=0.1)
y1 = np.sin(math.pi*x)
y2 = np.sin(math.pi*x + math.pi/2)
y3 = np.sin(math.pi*x - math.pi/2)

figure1 = plt.figure(figsize=[8,5])
axes1 = figure1.add_subplot(1,1,1)
axes1.plot(x,y1)
axes1.plot(x,y2)
axes1.plot(x,y3)
plt.show
```
