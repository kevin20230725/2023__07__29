```python
import numpy as np

student = np.array([67,88,96,74,55])
student[0]

#array只有索引編號，沒有key值
```

```python
import pandas as pd

s1 = pd.Series([67,88,96,74,55],index=['國','英','數','社','自'])
s1
```

```python
#Series 也能陣列運算
s1 + 3
```

```python
for i in s1:
    print(i)
```

```python
#numpy 的統計方法可以套用在pandas 上
import numpy as np
import pandas as pd
s1 = pd.Series([67,88,96,74,55],index=['國','英','數','社','自'])
print(s1)
print(f"加總為:{np.sum(s1)}")
print(f"平均為:{np.mean(s1)}")
print(f"標準差為:{np.std(s1)}")
print(f"中間值為:{np.median(s1)}")
```

```python
#Dataframe，由多個 Series 組成
import pandas as pd
pd.DataFrame({
    '陳楷仁':[70,80,90,75,85],
    '黃xx':[67,88,96,74,55],
    '劉xx':[99,55,72,65,93]
},index=['國','英','數','社','自'])
```
