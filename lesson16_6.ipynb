```python
import random
def get_score() -> list:
    score = []
    for i in range(5):
        score.append(random.randint(50,100))
    return score

#取得30個姓名
def get_names(num:int) -> list: 
    with open("names (1).txt",encoding="utf-8",newline='') as file:
        names_str = file.read()
        all_names_list = names_str.split("\n")      #將字串中的'\n'作為分割，形成一個list
        names_list = random.choices(all_names_list,k=num)
    return names_list

nums_list = int(input("請輸入學生數量:"))
#取得nums_list個名字
names_list = get_names(nums_list)
student = []
for i in range(nums_list):
    score = get_score()
    new_list = [names_list[i]] + score
    student.append(new_list)
student
```

```python
import pandas as pd
student_df = pd.DataFrame(student,columns=['姓名','國','英','數','社','自'],index=range(1,nums_list+1))
student_df
```

```python
student_df.index.name = '學號'
student_df.columns.name = '科目'
student_df
```

```python
sum_series = student_df['國'] + student_df['英'] + student_df['數'] + student_df['社'] + student_df['自']
sum_series
```

```python
student_df['總分'] = sum_series
student_df['平均'] = sum_series / 5.0
student_df
```

```python
#用 DataFrame 方法生成 csv  
student_df.to_csv("學生成績.csv")
```
