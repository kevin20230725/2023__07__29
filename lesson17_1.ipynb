```python
import pandas as pd
data = {'字串':['台北','台中'],
        "整數":[1000,3000],
        "浮點數":[3.258,7.396]}
df = pd.DataFrame(data)
display(df)
```

```python
#視覺美化
#需先下載 jinja2
#pip install jinja2
df.style.\
format(precision=2,thousands=',',decimal='.')\
.relabel_index(["列1",'列2'])
```

```python
import numpy as np
import pandas as pd
data = np.random.rand(10,2) * 5
df = pd.DataFrame(data,
             index = pd.date_range(start='20240324',periods=10),
             columns = ['台北','台中']
             )
df
```

```python
#pd.date_range() 函數測試
pd.date_range(start='2024-1-1',end='2024-1-10')
pd.date_range(start='2024-01-01', periods=10)
#以一周7天為單位，輸出每周最後一天的日期
pd.date_range(start='2024-01-01', end='2024-12-31', freq='W')
#以月為單位，輸出每月最後一天的日期
pd.date_range(start='2024-01-01', end='2024-12-31', freq='ME')
#以月為單位，輸出每月最後一個工作日的日期
pd.date_range(start='2024-01-01', end='2024-12-31', freq='BME')
#
pd.date_range(start='2024-01-01', end='2024-12-31', tz='UTC')

#pd.date_range() 日期參數有格式要求
#ISO 8601 标准日期格式：'YYYY-MM-DD'
#美国日期格式：'MM/DD/YYYY'
#欧洲日期格式：'DD/MM/YYYY'
#其他常见日期格式
```

```python
#視覺美化資料表
def feel_condition(v):
    if v < 1.75:
        return "溫暖"
    elif v < 2.75:
        return "潮濕"
    else :
        return "日曬"

df.style\
    .format(feel_condition)\
    .set_caption("體感溫度")\
    .background_gradient(cmap="PuBuGn",vmin=1,vmax=5,axis=None)
```

```python
#隱藏資料
df = pd.DataFrame(np.random.randn(5,5))
df.style.\
hide(subset=[0,2,4],axis=0)\
.hide(subset=[1,3],axis=1)
#將第1、3行和0、2、4列隱藏起來
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randn(10,4),columns=['A','B','C','D'])
display(df)

def style_negative(v, props=''):
    return props if v < 0 else None

df.style.map(style_negative, props='color:red')
```
