```python
display(student_frame['國'])
#如果用for in 迴圈，得到的output只有值，沒有索引
for item in student_frame['國']:
        if item>80:
            print('優')
        elif item>70:
              print('甲')
        elif item>70:
              print('乙')
        else :
              print('丙')
```

```python
#利用map()可以將output變成Series
def fun1(item):
    if item>80:
        return '優'
    elif item>70:
        return '甲'
    elif item>70:
        return '乙' 
    else :
        return '丙' 
student_frame['國'].map(fun1)
```

```python
student_frame['國文等級'] = student_frame['國'].map(fun1)
student_frame
```
