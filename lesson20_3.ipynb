```python
import yfinance as yf
data = yf.download("2330.TW", start='2024-01-01',end='2024-12-31')
tw2330 = data[['Adj Close']]
tw2330.columns = ['台積電']
tw2330
```

```python
import yfinance as yf
data = yf.download("2454.TW", start='2024-01-01',end='2024-12-31')
tw2454 = data[['Adj Close']]
tw2454.columns = ['聯發科']
tw2454
```

```python
import yfinance as yf
data = yf.download("2317.TW", start='2024-01-01',end='2024-12-31')
tw2317 = data[['Adj Close']]
tw2317.columns = ['鴻海']
tw2317
```

```python
import pandas as pd 
frame = pd.concat([tw2330,tw2317,tw2454],axis=1)
frame
```

```python
#顯示漲跌幅，df.pct_change()
frame1 = frame.pct_change()
frame2 = frame1.fillna(0)
frame2
```

```python
#彼此的線性關係，利用相關係數方法 corr()
frame2.corr()
```

```python
#漲幅最多
#frame2.max()
#跌幅最多
#frame2.min()
def fun1(s):
    max = s.max()*100
    min = s.min()*100
    return pd.Series([f'{max:.2f}%',f'{min:.2f}%'],index=['漲幅最多','跌幅最多'])

frame2.apply(fun1)
```
