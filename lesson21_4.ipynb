```python
import numpy as np
import pandas as pd
import requests

url1 = 'https://ods.railway.gov.tw/tra-ods-web/ods/download/dataResource/8ae4cabf6973990e0169947ed32454b9'
response = requests.get(url1)
jsonData = response.json()
jsonData
```

```python
dataFrame = pd.DataFrame(jsonData)
dataFrame.columns = ['乘車日','車站代碼','進站人數','出站人數']
dataFrame
```

```python
url2 = 'https://ods.railway.gov.tw/tra-ods-web/ods/download/dataResource/0518b833e8964d53bfea3f7691aea0ee'
response = requests.get(url2)
stations_json = response.json()
stations_json
```

```python
stations = pd.DataFrame(stations_json,columns=['stationCode','stationName'])
stations.columns = ['車站代碼','中文站名']
stations
```

21_5的merge後

```python
dataFrame1 = pd.merge(dataFrame,stations,on='車站代碼',how='left')
dataFrame1
```

```python
dataFrame1.info()
```

```python
dataFrame1['進站人數'] = dataFrame1['進站人數'].astype(int)
dataFrame1['出站人數'] = dataFrame1['出站人數'].astype(int)
```

```python
dataFrame1.info()
```

```python
#將乘車日轉成日期
dataFrame1['乘車日'] = pd.to_datetime(dataFrame1['乘車日'])
```

```python
dataFrame1.info()
```

```python
dataFrame2 = dataFrame1.set_index(keys=['乘車日','中文站名'])
dataFrame2
```

```python
#當日總進出站人數
dataFrame2.drop(columns=['車站代碼']).groupby(level='乘車日').sum()
```

```python
#每站總進出站人數
dataFrame2.drop(columns=['車站代碼']).groupby(level='中文站名').sum()
```
